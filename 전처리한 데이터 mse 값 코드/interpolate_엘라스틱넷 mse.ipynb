```python
import pandas as pd
data1 = pd.read_csv('C:/Users/pcy07/OneDrive/바탕 화면/2022 날씨공모전/interpolate_전체.csv')
data2 = pd.read_csv('C:/Users/pcy07/OneDrive/바탕 화면/2022 날씨공모전/interpolate_자외선제거.csv')
```

```python
data2.columns
```

```python
data1 = data1.drop(['Unnamed: 0', '날짜', '날짜_연도', '날짜_월', '날짜_일', '시간_분','경도','위도'], axis=1)
data2 = data2.drop(['Unnamed: 0', '날짜', '날짜_연도', '날짜_월', '날짜_일', '시간_분', '경도','위도'], axis=1)
```

## data1 : 전체 (자외선제거 X)

```python
# 지면타입/계절 원핫인코딩
data1_dum = pd.get_dummies(data1, columns = ['지면타입', '계절'])
data1_dum
```

```python
from sklearn.model_selection import train_test_split

X = data1_dum.drop(['자외선지수'], axis=1)
y = data1_dum['자외선지수']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
```

```python
from sklearn.linear_model import ElasticNet
import numpy as np

# 엘라스틱넷 적용 선형 회귀 모형 선언
## alpha는 규제의 강도를 의미하며, 높은 숫자일수록 패널티를 크게 부여함
model = ElasticNet(alpha = 0.01, l1_ratio = 0.5)
model1 = ElasticNet(alpha = 0.07, l1_ratio = 0.5)
model2 = ElasticNet(alpha = 0.1, l1_ratio = 0.5) 
model3 = ElasticNet(alpha = 0.5, l1_ratio = 0.5)
model4 = ElasticNet(alpha = 1, l1_ratio = 0.5)
```

```python
model.fit(X_train, y_train)
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

y_pred = model.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```

```python
y_pred = model1.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```

```python
y_pred = model2.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```

```python
y_pred = model3.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```

```python
y_pred = model4.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```
