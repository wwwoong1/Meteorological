```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/엘라스틱넷 돌릴 데이터.csv')
df2 = pd.read_csv('/content/drive/Shareddrives/날씨빅데이터티이임/mice 후 함수 적용_1.csv')
```

```python
del df1['Unnamed: 0']
del df2['Unnamed: 0']
```

엘라스틱 결측치 삭제 후

```python
df1
```

```python
# 변수마다 NaN의 비중
for col in df1.columns:
    na = 'column: {:>10}\t percent of NaN value: {:.2f}%'.format(col, 100 * (df1[col].isnull().sum() / df1[col].shape[0]))
    print(na)
```

```python
X = df1.drop(['자외선지수'], axis = 1)
y = df1['자외선지수']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=53)
```

```python
model1 = ElasticNet(alpha = 0.7, l1_ratio=0.5)
model2 = ElasticNet(alpha = 0.5, l1_ratio=0.5)
model3 = ElasticNet(alpha = 0.3, l1_ratio=0.5)
model4 = ElasticNet(alpha = 0.1, l1_ratio=0.5)
model5 = ElasticNet(alpha = 0.05, l1_ratio=0.5)
```

```python
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
model5.fit(X_train, y_train)
```

알파 = 0.7

```python
y_pred = model1.predict(X_test)
print('RMSE:', mean_squared_error(y_pred, y_test, squared = False))
print('r2_score: ', r2_score(y_test, y_pred))
```

알파 = 0.5

```python
y_pred = model2.predict(X_test)
print('RMSE:', mean_squared_error(y_pred, y_test, squared = False))
print('r2_score: ', r2_score(y_test, y_pred))
```

```python
y_pred = model3.predict(X_test)
print('RMSE:', mean_squared_error(y_pred, y_test, squared = False))
print('r2_score: ', r2_score(y_test, y_pred))
```

알파 = 0.1

```python
y_pred = model4.predict(X_test)
print('RMSE:', mean_squared_error(y_pred, y_test, squared = False))
print('r2_score: ', r2_score(y_test, y_pred))
```
