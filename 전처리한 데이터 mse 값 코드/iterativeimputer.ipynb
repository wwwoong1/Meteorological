결측치 대체 알고리즘을 실행

```python
df2 = df1.drop(['날짜','year','month','day'], axis = 1)
```

```python
df2 = pd.get_dummies(df2, columns = ['season'])
```

```python
df2
```

```python
df2 = df2.replace(-999, np.nan)
```

```python
df2
```

```python
from sklearn.experimental import enable_iterative_imputer
```

```python
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer(random_state=0)
imputer.fit(df2)
df3 = imputer.transform(df2)
df3
```

```python
ite_imputer = pd.DataFrame(df3)
```

```python
ite_imputer.columns = ['시간_분','관측소지점','위도','경도','자외선지수','파랑_가시밴드','초록_가시밴드','빨강_가시밴드','식생_가시밴드',
                       '눈/얼음_가시밴드','권운_밴드','야간안개/하층운_밴드','상층_수증기_밴드','중층_수증기_밴드','하층_수증기_밴드',
                       '구름상_밴드','오존_밴드','대기창_밴드','깨끗한_대기창_밴드','오염된_대기창_밴드','이산화탄소_밴드','태양천정각',
                       '위성천정각','대기외일사량','관측고도','지면타입','season_a','season_s','season_su','season_w']
```

```python
ite_imputer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X = ite_imputer.drop(['자외선지수'], axis = 1)
y = ite_imputer['자외선지수']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=53)
```

```python
model1 = ElasticNet(alpha = 0.7, l1_ratio=0.5)
model2 = ElasticNet(alpha = 0.5, l1_ratio=0.5)
model3 = ElasticNet(alpha = 0.3, l1_ratio=0.5)
model4 = ElasticNet(alpha = 0.1, l1_ratio=0.5)
model5 = ElasticNet(alpha = 0.05, l1_ratio=0.5)
```

```python
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
model5.fit(X_train, y_train)
```

0.7

```python
y_pred = model1.predict(X_test)
print('RMSE:', mean_squared_error(y_pred, y_test, squared = False))
print('r2_score: ', r2_score(y_test, y_pred))
```

```python
y_pred = model2.predict(X_test)
print('RMSE:', mean_squared_error(y_pred, y_test, squared = False))
print('r2_score: ', r2_score(y_test, y_pred))
```

0.3

```python
y_pred = model3.predict(X_test)
print('RMSE:', mean_squared_error(y_pred, y_test, squared = False))
print('r2_score: ', r2_score(y_test, y_pred))
```

```python
y_pred = model4.predict(X_test)
print('RMSE:', mean_squared_error(y_pred, y_test, squared = False))
print('r2_score: ', r2_score(y_test, y_pred))
```
