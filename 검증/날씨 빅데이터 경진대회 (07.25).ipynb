```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/Shareddrives/날씨빅데이터티이임/최종데이터.csv')
```

```python
df.info()
```

```python
df1 = df[(df['날짜_월']==5)|(df['날짜_월']==6)|(df['날짜_월']==7)]
```

```python
df1 = df1.drop(['날짜','날짜_연도','날짜_월','날짜_일','시간_분','경도','위도','위성천정각','관측고도','계절'], axis=1)
df1.reset_index(drop=True,inplace = True)
```

```python
df1.info()
```

```python
df1['Band1'] = df1.iloc[:,2:6].mean(axis=1)
df1['Band2'] = df1.iloc[:,6:8].mean(axis=1)
df1['Band3'] = df1.iloc[:,8:9].mean(axis=1)
df1['Band4'] = df1.iloc[:,9:12].mean(axis=1)
df1['Band5'] = df1.iloc[:,12:18].mean(axis=1)
```

```python
df1['태양천정각'].max()
```

```python
df1
```

```python
df1_1 = df1[(df1['태양천정각']>=0)&(df1['태양천정각']<=45)]
df1_1['time'] = 1
df1_2 = df1[(df1['태양천정각']>45)&(df1['태양천정각']<=90)]
df1_2['time'] = 2
df1_3 = df1[(df1['태양천정각']>90)&(df1['태양천정각']<=135)]
df1_3['time'] = 3
df1_4 = df1[(df1['태양천정각']>135)&(df1['태양천정각']<=180)]
df1_4['time'] = 4
```

```python
df_f = pd.concat([df1_1,df1_2,df1_3,df1_4])
```

```python
df_f = pd.get_dummies(df_f, columns = ['관측소지점'])
df_f = pd.get_dummies(df_f , columns = ['지면타입'])
df_f = pd.get_dummies(df_f, columns = ['time'])
```

```python
df_f.reset_index(drop=True,inplace = True)
```

```python
df_f
```

```python
from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_score, cross_validate,KFold #교차검증
from sklearn.model_selection import train_test_split

from sklearn.linear_model import RANSACRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
df_f.info()
```

```python
X = df_f.drop(['자외선지수','파랑_가시밴드','초록_가시밴드','빨강_가시밴드','식생_가시밴드','권운_밴드','눈/얼음_가시밴드',
              '야간안개/하층운_밴드','상층_수증기_밴드','중층_수증기_밴드','하층_수증기_밴드','구름상_밴드','오존_밴드',
              '대기창_밴드','깨끗한_대기창_밴드','오염된_대기창_밴드','이산화탄소_밴드'], axis=1)
y = df_f['자외선지수']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3,random_state=42)
```

```python
model =RANSACRegressor()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
k_fold = KFold(n_splits=10, shuffle=True)
scoring = {'R2 score' : r2_score(y_test, y_pred),
          'RMSE' : mean_squared_error(y_pred,y_test, squared=False)}
```
