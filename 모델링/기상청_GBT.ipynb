```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df1 = pd.read_csv('/content/drive/Shareddrives/날씨 빅데이터 분석 경진대회/모델링 데이터셋.csv')
df2 = pd.read_csv('/content/drive/Shareddrives/날씨 빅데이터 분석 경진대회/검증 데이터셋.csv')
```

```python
df1
```

```python
del df1['Unnamed: 0']
del df2['Unnamed: 0']
```

```python
X = df1.drop(['자외선지수'], axis=1)
y = df1['자외선지수']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 53)
```

```python
model = GradientBoostingRegressor(learning_rate = 0.01, random_state=42, n_estimators = 500, max_depth = 5, subsample = 1)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
score1 = {'R2_score' : r2_score(y_test, y_pred)}
score2 = {'RMSE' : mean_squared_error(y_pred, y_test, squared = False)}
score3 = {'MSE' : mean_absolute_error(y_pred, y_test)}
```

```python
score1
```

```python
test = test.drop(['자외선지수'], axis=1)
```
