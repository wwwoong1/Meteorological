## 엘라스틱넷

```python
from sklearn.model_selection import train_test_split

X = data_dum.drop(['자외선지수'], axis=1)
y = data_dum['자외선지수']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 99)
```

```python
from sklearn.linear_model import ElasticNet
import numpy as np

# 엘라스틱넷 적용 선형 회귀 모형 선언
## alpha는 규제의 강도를 의미하며, 높은 숫자일수록 패널티를 크게 부여함
model1 = ElasticNet(alpha = 0.0001, l1_ratio = 0.5)
model2 = ElasticNet(alpha = 0.001, l1_ratio = 0.5)
model3 = ElasticNet(alpha = 0.01, l1_ratio = 0.5)
model4 = ElasticNet(alpha = 0.1, l1_ratio = 0.5)
```

```python
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# alpha = 0.0001
y_pred = model1.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```

```python
# alpha = 0.001
y_pred = model2.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```

```python
# alpha = 0.01
y_pred = model3.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```

```python
# alpha = 0.1
y_pred = model4.predict(X_test) # 예측 값
print(mean_squared_error(y_pred, y_test, squared = False))
print(r2_score(y_test,y_pred))
```
