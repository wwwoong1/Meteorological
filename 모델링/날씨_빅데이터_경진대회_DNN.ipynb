```python
df1_1 = df1[(df1['태양천정각']>=0)&(df1['태양천정각']<=45)]
df1_1['time'] = 1
df1_2 = df1[(df1['태양천정각']>45)&(df1['태양천정각']<=90)]
df1_2['time'] = 2
df1_3 = df1[(df1['태양천정각']>90)&(df1['태양천정각']<=135)]
df1_3['time'] = 3
df1_4 = df1[(df1['태양천정각']>135)&(df1['태양천정각']<=180)]
df1_4['time'] = 4
```

```python
X = df_f.drop(['자외선지수','파랑_가시밴드','초록_가시밴드','빨강_가시밴드','식생_가시밴드','권운_밴드','눈/얼음_가시밴드',
              '야간안개/하층운_밴드','상층_수증기_밴드','중층_수증기_밴드','하층_수증기_밴드','구름상_밴드','오존_밴드',
              '대기창_밴드','깨끗한_대기창_밴드','오염된_대기창_밴드','이산화탄소_밴드'], axis=1)
y = df_f['자외선지수']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3,random_state=42)
```

```python
model =RANSACRegressor()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
k_fold = KFold(n_splits=10, shuffle=True)
scoring = {'R2 score' : r2_score(y_test, y_pred),
          'RMSE' : mean_squared_error(y_pred,y_test, squared=False)}
```

```python
import tensorflow as tf
import tensorflow.keras as keras
```

```python
np.random.seed(1)
tf.random.set_seed(1)

learning_rate = 0.001
n_epochs = 100
n_batch = 32
```

```python
train_data = tf.data.Dataset.from_tensor_slices((X_train, y_train)).shuffle(500).batch(n_batch, drop_remainder = True).repeat()
```

```python
test_data = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(n_batch)
```

```python
def create_model():
  model = keras.Sequential()
  model.add(keras.layers.Dense(units=16, activation = 'relu', input_shape = (29,)))
  model.add(keras.layers.Dense(units=8, activation = 'relu'))
  model.add(keras.layers.Dense(units=4, activation = 'relu'))
  model.add(keras.layers.Dense(units=1))

  return model
```

```python
model = create_model()
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate = learning_rate), loss = 'mse')
```

```python
model.summary()
```

```python
steps_per_epoch = X_train.shape[0] // n_batch
```

```python
validation_steps = int(np.ceil(X_test.shape[0]/n_batch))
```

```python
from keras.callbacks import EarlyStopping
callback = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 10)
```

```python
history = model.fit(train_data,epochs = n_epochs,
          steps_per_epoch = steps_per_epoch,
          validation_data = test_data,
          validation_steps = validation_steps)
```
