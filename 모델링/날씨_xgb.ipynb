```python
df = df.drop(['Unnamed: 0','날짜','날짜_연도','날짜_일','시간_분','경도','위도','위성천정각','계절'], axis = 1)
```

```python
df1 = df[(df['날짜_월']==5)|(df['날짜_월']==6)|(df['날짜_월']==7)]
```

```python
df1.reset_index(drop=True, inplace = True)
```

```python
df1
```

```python
df1['Band1'] = df1.iloc[:,2:6].mean(axis=1)
df1['Band2'] = df1.iloc[:,6:8].mean(axis=1)
df1['Band3'] = df1.iloc[:,8:9].mean(axis=1)
df1['Band4'] = df1.iloc[:,9:12].mean(axis=1)
df1['Band5'] = df1.iloc[:,12:18].mean(axis=1)
```

```python
df1_1 = df1[(df1['태양천정각']>=0)&(df1['태양천정각']<=45)]
df1_1['time'] = 1
df1_2 = df1[(df1['태양천정각']>45)&(df1['태양천정각']<=90)]
df1_2['time'] = 2
df1_3 = df1[(df1['태양천정각']>90)&(df1['태양천정각']<=135)]
df1_3['time'] = 3
df1_4 = df1[(df1['태양천정각']>135)&(df1['태양천정각']<=180)]
df1_4['time'] = 4
```

```python
data = pd.concat([df1_1,df1_2,df1_3,df1_4])
```

```python
data = pd.get_dummies(data, columns = ['관측소지점'])
data = pd.get_dummies(data , columns = ['지면타입'])
data = pd.get_dummies(data, columns = ['time'])
```

```python
data.reset_index(drop=True,inplace = True)
```

```python
data
```

```python
from sklearn.model_selection import cross_val_score, cross_validate,KFold,GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X = data.drop(['자외선지수','파랑_가시밴드','초록_가시밴드','빨강_가시밴드','식생_가시밴드','권운_밴드','눈/얼음_가시밴드',
              '야간안개/하층운_밴드','상층_수증기_밴드','중층_수증기_밴드','하층_수증기_밴드','구름상_밴드','오존_밴드',
              '대기창_밴드','깨끗한_대기창_밴드','오염된_대기창_밴드','이산화탄소_밴드'], axis=1)
y = data['자외선지수']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3,random_state=42)
```

```python
import xgboost
```

```python
model = xgboost.XGBRegressor()
```

```python
params = {
    'n_estimators':(100, 150),
    'max_depth' : (5, 8),
    'min_samples_leaf' : (8, 18),
    'min_samples_split' : (8, 16)
}
```

```python
grid_cv = GridSearchCV(xgboost.XGBRegressor(random_state=42), params, n_jobs=-1)
grid_cv.fit(X_train, y_train)
```

```python
print(grid_cv.best_params_)
```

```python
best_gd = grid_cv.best_estimator_
```

```python
xgb_g = grid_cv.best_estimator_
print(xgb_g.score(X_test, y_test))
```

```python
# test
y_preds1 = xgb_g.predict(X_test)
mse = mean_squared_error(y_test, y_preds1)
rmse = np.sqrt(mse)
r2_points = r2_score(y_test, y_preds1)
print("RMSE:", rmse)
print("R^2:", r2_points)
```

```python
# train
y_preds2 = xgb_g.predict(X_train)
mse = mean_squared_error(y_train, y_preds2)
rmse = np.sqrt(mse)
r2_points = r2_score(y_train, y_preds2)
print("RMSE:", rmse)
print("R^2:", r2_points)
```
